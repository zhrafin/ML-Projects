```python
import pandas as pd
import numpy as np
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for file in os.listdir(r'C:\Users\rafin\t20s'):
    filenames.append(os.path.join(r'C:\Users\rafin\t20s', file))
```

```python
filenames[:5]
```

```python
dfs = []

# Process each file and collect the dataframes
for counter, file in enumerate(tqdm(filenames), start=1):
    try:
        with open(file, 'r') as f:
            df = pd.json_normalize(safe_load(f))
            df['match_id'] = counter
            dfs.append(df)
    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Concatenate all collected dataframes into the final dataframe
final_df = pd.concat(dfs, ignore_index=True)
```
